```python
cat = []
num = []
for i in df_cleaned.columns:
    if df_cleaned[i].dtypes == 'O':
        cat.append(i)
    else:
        num.append(i)
```

```python
Categorical_df = df_cleaned[cat]
Numerical_df = df_cleaned[num]

Categorical_df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
labencoder = {}
cat_cols = Categorical_df.columns

for column in cat_cols:
    labencoder[column] = LabelEncoder()
    Categorical_df[column] = labencoder[column].fit_transform(Categorical_df[column])

finalFrames = [Categorical_df,Numerical_df]
df_encoded = pd.concat(finalFrames,axis=1)

df_encoded.info()
```

```python
df_featuredSelected=df_encoded[["highwaympg","citympg","CarName",'enginelocation','fuelsystem','boreratio','wheelbase','drivewheel','carlength','carwidth','horsepower','curbweight','enginesize','price']]
```

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X = df_featuredSelected.drop(columns="price")
y = df_featuredSelected["price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=45)

LR = LinearRegression()

LR.fit(X_train,y_train)
y_test_pred = LR.predict(X_test)
```

```python
import numpy as np
RSquare_Value = r2_score(y_test,y_test_pred)
RMSE_Value = np.sqrt(mean_squared_error(y_test,y_test_pred))

print(RSquare_Value)
print(RMSE_Value)

print(f"Intercept: {LR.intercept_}")
print(f"Coefficients: {LR.coef_}")
```
